```python


def minimize_energy_spread_dft(x, fs, f1, f2, floor_db=-120.0):
   
    T1 = int(round(fs / f1))
    T2 = int(round(fs / f2))

    M = (T1 * T2) // gcd(T1, T2)

    if M % 2 == 1:
        M *= 2

    xM = x[:M]
    X = fft(xM, n=M)
    hM = M // 2
    mX = 20 * np.log10(np.maximum(np.abs(X[:hM + 1]), 1e-12))  
    mX[mX < floor_db] = floor_db  
    return mX, M
```
